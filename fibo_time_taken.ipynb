```python
import matplotlib.pyplot as plt
from datetime import datetime
n= int(input("entr no"))
l1=[]
l2=[]
a=0
b=1
for i in range(0,n):
    l1.append(a)
    t1=datetime.now()
    c=a
    a=b
    b=c+b
    t2=datetime.now()
    e=t2.microsecond-t1.microsecond
    l2.append(e)
print("fibbonaci series",l1)
print("time taken",l2)


plt.plot(l1,l2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import time


def fibonacci(n):
    if(n <= 1):
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))
n = int(input("Enter number of terms:"))
print("Fibonacci sequence:")
a1=[]
a2=[]
for i in range(n):
    t3 = time.time()
    g=fibonacci(i)
    t4=time.time()
    f=t4-t3
    a1.append(g)
    a2.append(f)
print(a1)
print(a2)


plt.plot(a1,a2)
plt.show()
```
